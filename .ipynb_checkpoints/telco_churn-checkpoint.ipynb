```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("telco_churn.csv")

print("Dimensi Data: ", df.shape)

# Audit tipe data dan deteksi memori
print("\n--- Metadata ---")
df.info()

# Pratinjau 5 baris pertama untuk memvalidasi struktur
df.head()
```

```python
# Menimpa nilai kosong dengan angka 0 secara permanen
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Validasi ulang untuk memastikan integritas matriks
print("Sisa nilai kosong di TotalCharges:", df['TotalCharges'].isnull().sum())
```

```python
# 1. Konversi string spasi menjadi representasi null komputasional (NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# 2. Filter data untuk melihat HANYA baris yang nilai TotalCharges-nya kosong
anomali_tagihan = df[df['TotalCharges'].isnull()]

# 3. Inspeksi relasi antara nilai kosong dengan metrik 'tenure' (lama berlangganan)
print(anomali_tagihan[['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']])
```
